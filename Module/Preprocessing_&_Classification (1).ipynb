```python
import json
import re
import nltk
import spacy
from nltk.corpus import stopwords
from collections import defaultdict

# Download stopwords
nltk.download('stopwords')
nlp = spacy.load("en_core_web_sm")

# Define extracted folder (Source Data)
extracted_folder = "/content/drive/MyDrive/Extracted Dataset from PDF"

# âœ… Define Optimized Folder with Structured Subfolders
optimized_output_folder = "/content/drive/MyDrive/Optimized_Preprocessed_Text"

structured_folders = {
    "Civil_Train": os.path.join(extracted_folder, "Civil", "Train"),
    "Civil_Validation": os.path.join(extracted_folder, "Civil", "Validation"),
    "Civil_Testing": os.path.join(extracted_folder, "Civil", "Testing"),
    "Criminal_Train": os.path.join(extracted_folder, "Criminal", "Train"),
    "Criminal_Validation": os.path.join(extracted_folder, "Criminal", "Validation"),
    "Criminal_Testing": os.path.join(extracted_folder, "Criminal", "Testing"),
}

# âœ… Create Optimized Subfolders in the Same Structure
optimized_folders = {}
for key, path in structured_folders.items():
    new_path = path.replace(extracted_folder, optimized_output_folder)
    os.makedirs(new_path, exist_ok=True)
    optimized_folders[key] = new_path  # Store optimized folder paths

# Load stopwords
stop_words = set(stopwords.words('english'))

# Case Type Identification
case_type_keywords = {
    "civil": ["contract", "property", "divorce", "alimony", "tenant", "landlord", "dispute", "claims"],
    "criminal": ["murder", "assault", "kidnapping", "theft", "forgery", "robbery", "fraud", "crime", "cheating"]
}

# âœ… Function to Extract Metadata
def extract_metadata(text):
    court_name_match = re.search(r'(High Court|District Court|Sessions Court|Supreme Court|Tribunal|Bench)', text, re.IGNORECASE)
    court_name = court_name_match.group(1) if court_name_match else "Not Available"

    case_type = "Unknown"
    for type_name, keywords in case_type_keywords.items():
        if any(re.search(rf'\b{keyword}\b', text, re.IGNORECASE) for keyword in keywords):
            case_type = type_name.capitalize()
            break

    acts = list(set(re.findall(r'\b[A-Z][a-z]+ Act\b', text)))
    laws = list(set(re.findall(r'\b[A-Z][a-z]+ Law\b', text)))
    all_acts_laws = list(set(acts + laws)) if (acts or laws) else ["Not Available"]

    return {
        "court_name": court_name,
        "case_type": case_type,
        "acts": all_acts_laws
    }

# âœ… Function to Split Text into Sentences
def split_into_sentences(text):
    return re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s+', text)

# âœ… Function to Extract Arguments & Decisions (With Deduplication)
argument_keywords = {
    "argued", "submits", "contended", "prosecution", "defendant", "respondent",
    "appellant", "the learned counsel", "the petitioner states", "the prosecution argued",
    "the appellant argued", "submissions made by", "it was contended that",
    "the defense counsel contended", "it is the submission of"
}

decision_keywords = {
    "judgment", "dismissed", "decision", "ordered", "decree", "granted",
    "allowed", "denied", "it is ordered that", "the appeal is dismissed",
    "we are of the considered opinion", "this court directs", "the court held that",
    "it is adjudged", "the final ruling is", "case is resolved", "we find that"
}

def extract_unique_context(text, keywords):
    sentences = split_into_sentences(text)
    extracted = defaultdict(set)  # âœ… Use a set to store unique values

    for sentence in sentences:
        for keyword in keywords:
            if keyword in sentence.lower():
                extracted[keyword].add(sentence.strip())  # âœ… Automatically removes duplicates

    return {key: list(value) for key, value in extracted.items()}  # Convert set back to list for JSON storage

# âœ… Function to Process Each Folder
def process_folder(folder_path, output_folder):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)

        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read()

                # ğŸ”¹ Extract Metadata
                metadata = extract_metadata(text)

                # ğŸ”¹ Extract Arguments & Decisions (With Deduplication)
                argument_contexts = extract_unique_context(text, argument_keywords)
                decision_contexts = extract_unique_context(text, decision_keywords)

                # ğŸ”¹ Handle Missing Arguments/Decisions
                if not argument_contexts:
                    argument_contexts["Auto-Extracted"] = ["No clear arguments found. Needs manual review."]
                if not decision_contexts:
                    last_100_lines = text.split("\n")[-100:]
                    decision_contexts["Auto-Extracted"] = last_100_lines
                    print(f"âš ï¸ Decision section missing! Extracted from last 100 lines for {file_name}.")

                # âœ… Format Output Data (Without Common Phrases)
                output_data = {
                    "file_name": file_name,

                    "meta_data": {
                        "court_name": metadata["court_name"],
                        "case_type": metadata["case_type"],
                        "acts": metadata["acts"]
                    },

                    "separator_1": "---------------------- FACTS ----------------------\n\n",
                    "facts": text.strip(),

                    "separator_2": "\n\n---------------------- ARGUMENTS ----------------------\n\n",
                    "arguments": argument_contexts,

                    "separator_3": "\n\n---------------------- DECISIONS ----------------------\n\n",
                    "decisions": decision_contexts
                }

                # âœ… Remove ".txt" from the filename before saving
                cleaned_file_name = file_name.replace(".txt", "")

                # âœ… Save Fine-Tuned Data in JSON Format
                output_path = os.path.join(output_folder, f"{cleaned_file_name}.json")
                with open(output_path, "w", encoding="utf-8") as f:
                    json.dump(output_data, f, indent=4)

                print(f"âœ… Processed & Saved: {output_path}")

# âœ… Process Each Folder
for key, path in structured_folders.items():
    process_folder(path, optimized_folders[key])

print("\nğŸ‰ All preprocessed files have been saved in 'Optimized_Preprocessed_Text' with NO '.txt' in filenames! ğŸš€")
```
